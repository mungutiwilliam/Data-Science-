```python
data
```

the concept of hot encoding binary representation is used for all the features expect the name 
Hot encoding is using '0' and '1'
We have a multiclass kind of problem but with very limited data therefore we convert the problem to a binary problem

```python
data['class']= data['class'].replace(['Fish','Bird','Amphibian','Reptile'], 'non-mammals')
```

```python
data
```

```python
data.drop('Name', axis=1)
```

the data has now been made to a binary problem with mammals and non-mammals

pandas cross tabulation can be used to show the relationship between warm blooded and gives birth

```python
pd.crosstab([data['Warmblooded'],data['gives birth']], data['class'])
```

# Building the Decision Tree classifier

```python
from sklearn import tree
```

```python
y=data['class']
```

```python
y
```
